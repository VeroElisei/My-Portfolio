Advanced Build-In Functions

```python
#Map -> takes a function and a sequence as inputs; returns an iterator that applies the function to every item in the sequence

import pandas as pd
```

```python
grades = pd.read_csv("../input/grades/grades.csv")
grades.head()
```

```python
def lowest_grade(student_id):
    return grades.loc[grades["student_id"] == student_id]["grade"].fillna(0).min()

```

```python
assert lowest_grade(1) == 0.0, "test failed"
print("test passed")
```

```python
#sequence containing all distinct students ids
student_ids = grades["student_id"].unique()
student_ids
```

```python
list(map(lowest_grade, student_ids))
```

```python
#Zip -> takes two or more sequences as input; allows to iterate through all the sequences at the same time

points_game1 = [50, 40, 60, 70, 80]
points_game2 = [76, 81, 53, 92, 67]
#sequence where points differences will be saved
diffs = []

#iterate through points_game1 and points_game2 at the same time
for x, y in zip(points_game1, points_game2):
    diffs.append(abs(x-y))

diffs
```

```python
#Filter
def mean_atleast_70(student_id):
    mean_grade = grades.loc[grades["student_id"] == student_id]["grade"].fillna(0).mean()
    return mean_grade >= 70
```

```python
assert mean_atleast_70(1) == False, "test failed"
print("test passed")
```

```python
list(filter(mean_atleast_70, student_ids))
```

#Numpy arrays

One-dimentional numpy arrays are considered vectors


Multi-dimentional numpy arrays are considered matrices

```python
import numpy as np
```

```python
#Create an empty array:
array0 = np.array([])
array0 
```

```python
#create 1D array
list1 = [1,2,3,4,5]
array1 = np.array(list1)
array1
```

```python
#create 2D array
list2 = [[1,2,3], [4,5, 6], [7,8,9]]
array2 = np.array(list2)
array2
```

```python
#create arrays from an interval; 3 numerical inputs: start, stop (excluded), step (optional)
x = np.arange(5)
y = np.arange(0,11,2)
print(x)
print(y)
```

```python
#linspace function:  start, stop (included), num(how many no to generate)
np.linspace(1,2,9)
```

```python
#random.randin: low, high (exclusive), size (how many)
x = np.random.randint(20,50,10)
print(x)
```

```python
#min value from array
x.min()
```

```python
#max value from array
x.max()
```

```python
#index of min / max values
array_random = np.random.randint(1,50,20)
array_random
x = array_random.argmin()
y = array_random.argmax()
z = array_random.min()
t = array_random.max()
print(array_random)
print(x) # -> index of the min value
print(y) # -> index of the max value
print(z) #min value
print(t) #max value
```

```python
#reshape an array:

array_random = np.random.randint(2,20,10)
array_random
b = array_random.reshape(5,2)
print(array_random)
print("\n")
print(b)
```

```python
#conditional selection for arrays:
array = np.arange(0,21,2)
print(array)
print(array[(array >6) & (array <15)])
```
